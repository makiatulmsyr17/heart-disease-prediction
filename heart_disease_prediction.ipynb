# Understanding Data

Variable Explanation

1. **age**: This variable indicates the age of the patient in years.

2. **sex**: This variable indicates the gender of the patient, where a value of 1 represents a male and a value of 0 represents a female.

3. **cp**: This variable describes the type of chest pain the patient is experiencing. Its value can range between 0 and 3, with each value representing a specific type of pain.
Value 0: typical angina
Value 1: atypical angina
Value 2: non-anginal pain
Value 3: asymptomatic



4. **trestbps**: This is the patient's blood pressure at rest, measured in millimeters of mercury (mmHg) on admission to the hospital.
Normal: Systolic blood pressure less than 120 mmHg and diastolic blood pressure less than 80 mmHg.
Prehypertension: Systolic blood pressure 120-139 mmHg or diastolic blood pressure 80-89 mmHg.
Stage 1 hypertension: Systolic blood pressure 140-159 mmHg or diastolic blood pressure 90-99 mmHg.
Stage 2 hypertension: Systolic blood pressure 160 mmHg or higher or diastolic blood pressure 100 mmHg or higher.

5. **chol**: Represents the patient's total cholesterol level, measured in milligrams per deciliter (mg/dL).
Less than 200mg/dL Desirable
200-239 mg/dL Borderline high
240mg/dL and above High



6. **fbs**: Indicates the patient's fasting blood sugar level. If the value is 1, it indicates that the fasting blood sugar level exceeds 120 mg/dL; if the value is 0, it indicates that the fasting blood sugar level is at or below 120 mg/dL.
Considered prediabetes


7. **restecg**: This variable records the patient's resting electrocardiogram (ECG) results, with values between 0 and 2.
Value 0: normal
Value 1: has ST-T wave abnormalities (T wave inversion and/or ST elevation or depression > 0.05 mV)
Score 2: indicates probable or definite left ventricular hypertrophy according to Estes criteria

8. **thalach**: This is the maximum heart rate the patient achieved during the exercise test. maximum heart rate achieved: MHR = 220 - age

9. **exang**: Indicates whether the patient experienced exercise-induced angina. The value is 1 if the patient has exercise-induced angina and 0 otherwise.

10. **oldpeak**: Represents the ST segment depression caused by exercise relative to rest.

11. **slope**: This variable records the slope of the exercise ST segment, with values between 0 and 2.
Value 0: upsloping
Value 1: flat
Value 2: downsloping

12. **ca**: Indicates the number of major blood vessels (0-3) recorded as having more than 50% narrowing.

13. **thal**: This variable describes the type of heart defect observed. The value can be 0 (normal), 1 (fixed defect), or 2 (reversible defect).

14. **target**: This target variable indicates the status of the patient's heart disease diagnosis. Its value is 1 if the patient is diagnosed with heart disease and 0 otherwise.

We will categorize the trestbps column (blood pressure at rest)
 blood pressure at rest:

- Normal: Systolic blood pressure less than 120 mmHg and diastolic blood pressure less than 80 mmHg.
- Prehypertension: Systolic blood pressure 120-139 mmHg or diastolic blood pressure 80-89 mmHg.
- Stage 1 hypertension: Systolic blood pressure of 140-159 mmHg or diastolic blood pressure of 90-99 mmHg.
- Stage 2 hypertension: Systolic blood pressure of 160 mmHg or higher or diastolic blood pressure of 100 mmHg or higher.


```python
def categorize_blood_pressure(trestbps):
    if trestbps < 120:
        return 0  # Normal
    elif 120 <= trestbps <= 139:
        return 1  # Prehypertension
    elif 140 <= trestbps <= 159:
        return 2  # Stage 1 Hypertension
    else:
        return 3  # Stage 2 Hypertension


# Assuming 'df' is your DataFrame containing the data
df_clean["trestbps_category"] = df_clean["trestbps"].apply(categorize_blood_pressure)
df_clean
```

We will also categorize the chol column (the patient's total cholesterol level).
Serum cholesterol is measured in milligrams per deciliter (mg/dL). The normal range for serum cholesterol is:

- Less than 200mg/dL Desirable
- 200-239 mg/dL Borderline high
- 240mg/dL and above High


```python
def categorize_chol(value):
    if value < 200:
        return 0  # Desirable
    elif value < 240:
        return 1  # Borderline high
    else:
        return 2  # High


df_clean["chol_category"] = df_clean["chol"].apply(categorize_chol)
df_clean
```

(thalach) maximum heart rate achieved
The formula for getting the maximum heart rate is 220 minus age in minutes. so we will create a column using that formula.

```python
# Fungsi untuk menghitung Max Heart Rate
def mhRate(age):
    return 220 - age

# Menggunakan fungsi mhRate dengan metode apply untuk menghitung MHR untuk setiap umur
df_clean["mhr"] = df_clean["age"].apply(mhRate)


df_clean = pd.DataFrame(df_clean)

# Menampilkan DataFrame df yang sudah dimodifikasi
print(df_clean)
```

Here we will create a category that compares the patient's heart rate in the thalach and mhr columns that we have calculated and created columns for, and will create a category column.

```python
# Definisikan fungsi mhrCat (0 = bagus 1 = tidak bagus  )
def mhrCat(talach, mhr):
    return 0 if talach <= mhr else 1

# Tambahkan kolom "heart_rate_category" ke DataFrame df dengan menggunakan fungsi apply dan lambda
df_clean["heart_rate_category"] = df_clean.apply(lambda row: mhrCat(row["thalach"], row["mhr"]), axis=1)

# Menampilkan DataFrame df yang sudah dimodifikasi
print(df_clean)
```

### Exploring Additional Models

```python
# Exploring Additional Models

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC


# 1. Logic regretion
log = LogisticRegression(random_state=4, solver = "newton-cg")
log.fit(x_train, y_train)
pred = log.predict(x_test)
score_lg_train = log.score(x_train, y_train)
print("LogisticRegression Accuracy train", score_lg_train)
score_lg_test = log.score(x_test, y_test)
print("LogisticRegression Accuracy test:", score_lg_test)

# 2. Model Decision Tree
model_dt = DecisionTreeClassifier(random_state=0)
model_dt.fit(x_train, y_train)
score_dt = model_dt.score(x_test, y_test)
print("Decision Tree Accuracy:", score_dt)

# 3. Model Random Forest
model_rf = RandomForestClassifier(random_state=0)
model_rf.fit(x_train, y_train)
score_rf = model_rf.score(x_test, y_test)
print("Random Forest Accuracy:", score_rf)

# 4. Model Support Vector Machine (SVM)
model_svm = SVC(random_state=0)
model_svm.fit(x_train, y_train)
score_svm = model_svm.score(x_test, y_test)
print("SVM Accuracy:", score_svm)
```

```python
from sklearn.model_selection import cross_val_score

# 1. Cross-validation untuk Logistic Regression
cv_scores_log = cross_val_score(log, x_train, y_train, cv=5)
print("Cross-validation scores for Logistic Regression:", cv_scores_log)
print("Average cross-validation score for Logistic Regression:", cv_scores_log.mean())

# 2. Cross-validation untuk Decision Tree
cv_scores_dt = cross_val_score(model_dt, x_train, y_train, cv=5)
print("Cross-validation scores for Decision Tree:", cv_scores_dt)
print("Average cross-validation score for Decision Tree:", cv_scores_dt.mean())

# 3. Cross-validation untuk Random Forest
cv_scores_rf = cross_val_score(model_rf, x_train, y_train, cv=5)
print("Cross-validation scores for Random Forest:", cv_scores_rf)
print("Average cross-validation score for Random Forest:", cv_scores_rf.mean())

# 4. Cross-validation untuk Support Vector Machine (SVM)
cv_scores_svm = cross_val_score(model_svm, x_train, y_train, cv=5)
print("Cross-validation scores for SVM:", cv_scores_svm)
print("Average cross-validation score for SVM:", cv_scores_svm.mean())
```

Hitung rata-rata skor dari cross-validation untuk setiap model

```python
# Hitung rata-rata skor dari cross-validation untuk setiap model
avg_scores = {
    'Logistic Regression': np.mean(cross_val_score(log, x_train, y_train, cv=5)),
    'Decision Tree': np.mean(cross_val_score(model_dt, x_train, y_train, cv=5)),
    'SVM': np.mean(cross_val_score(model_svm, x_train, y_train, cv=5))
}

# Tampilkan hasil kesimpulan
for model, score in avg_scores.items():
    print(f"Average Cross-Validation Score for {model}: {score}")

# Pilih model dengan skor tertinggi
best_model = max(avg_scores, key=avg_scores.get)
print(f"\nThe best model is {best_model} with an average cross-validation score of {avg_scores[best_model]}")
```

```python
# Kita ambil yg Logic regretion

from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# Menghitung akurasi pada data pelatihan
accuracy = log.score(x_train, y_train)

# Memprediksi label menggunakan data pengujian
y_pred = log.predict(x_test)

# Menghitung confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Menghitung metrik evaluasi lainnya
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Menampilkan hasil evaluasi
print("Confusion Matrix:")
print(conf_matrix)
print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Membuat kesimpulan
if accuracy > 0.8 and precision > 0.8 and recall > 0.8 and f1 > 0.8:
    print("\nModel memiliki kinerja yang sangat baik.")
elif accuracy > 0.7 and precision > 0.7 and recall > 0.7 and f1 > 0.7:
    print("\nModel memiliki kinerja yang baik.")
else:
    print("\nModel memiliki kinerja yang perlu diperbaiki.")
```

- Accuracy: Indicates how well the model can predict overall. If the accuracy is high, it indicates that the model is likely to have good performance.

- Precision: Indicates how well the model can predict the positive class without classifying the negative as positive. High precision indicates that the model rarely gives false positives.

- Recall: Indicates how well the model can find all instances of the positive class. A high recall indicates that the model rarely misses true positive class instances.

- F1-score: Is the harmonic mean of precision and recall, giving an overall picture of the model's performance.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat heatmap dari confusion matrix
sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d", cbar=False)

# Menambahkan label sumbu x dan y
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

# Menampilkan plot
plt.show()
```

In the context of predicting heart disease, we want the model to perform well in identifying patients who actually have heart disease (positive) and patients who are healthy (negative). Let's evaluate the confusion matrix:

- True Positives (TP): 33
The number of patients who actually had heart disease and were correctly predicted by the model.
- True Negatives (TN): 15
The number of patients who are completely healthy and correctly predicted by the model.
- False Positives (FP): 11
The number of patients who were actually healthy, but were incorrectly predicted as having heart disease by the model.
- False Negatives (FN): 2
The number of patients who actually had heart disease, but were incorrectly predicted as healthy by the model.

```python
# Mengatur nilai threshold untuk klasifikasi hasil prediksi dapat membantu mengurangi salah satu jenis kesalahan.
y_pred_proba = log.predict_proba(x_test)[:, 1]

# Mengubah threshold
threshold = 0.3 # contoh nilai threshold baru
y_pred_adjusted = (y_pred_proba > threshold).astype(int)

# Evaluasi model dengan threshold yang disesuaikan
conf_matrix_adjusted = confusion_matrix(y_test, y_pred_adjusted)
precision_adjusted = precision_score(y_test, y_pred_adjusted)
recall_adjusted = recall_score(y_test, y_pred_adjusted)
f1_adjusted = f1_score(y_test, y_pred_adjusted)

print("Confusion Matrix (Threshold Adjusted):")
print(conf_matrix_adjusted)
print("\nPrecision (Threshold Adjusted):", precision_adjusted)
print("Recall (Threshold Adjusted):", recall_adjusted)
print("F1 Score (Threshold Adjusted):", f1_adjusted)

# Membuat kesimpulan
if precision_adjusted > 0.8 and recall_adjusted > 0.8 and f1_adjusted > 0.8:
    print("Model memiliki kinerja yang sangat baik.")
elif precision_adjusted > 0.7 and recall_adjusted > 0.7 and f1_adjusted > 0.7:
    print("Model memiliki kinerja yang baik.")
else:
    print("Model memiliki kinerja yang perlu diperbaiki.")
```

- Setting the threshold value for classification of predicted results can help reduce either type of error. For example, by lowering the threshold, we can reduce False Negatives (FN) but increase False Positives (FP), and vice versa. This depends on the specific preferences and needs in the clinical application.
- By lowering the threshold from 0.5 to 0.3, we can reduce False Negatives (FN), but may increase False Positives (FP).
